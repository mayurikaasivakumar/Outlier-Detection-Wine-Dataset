```python
wine=load_wine()
X=wine.data
display(X)
```

```python
from sklearn.ensemble import IsolationForest
clf=IsolationForest(n_estimators=100, contamination=0.5, random_state=42)
clf.fit(X)
y_pred = clf.predict(X)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='copper')
plt.title("Isolation Forest Outlier Detection on WINE Dataset")
plt.xlabel("Alcohol Content ")
plt.ylabel("pH Level")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
```

```python
# Load the WINE dataset
wine = load_wine()
X = wine.data
y = wine.target
```

```python
# Fit Isolation Forest model
clf_iso = IsolationForest(contamination=0.5, random_state=42)
y_pred_iso = clf_iso.fit_predict(X)
```

```python
# Fit Local Outlier Factor model
clf_lof = LocalOutlierFactor(contamination=0.5)
y_pred_lof = clf_lof.fit_predict(X)
```

```python
# Plot Isolation Forest outliers
plt.scatter(X[:, 0], X[:, 1], c=np.where(y_pred_iso == -1, 'pink', 'purple'), label='Isolation Forest')
plt.title("Outlier Detection using Isolation Forest on WINE Dataset")
plt.xlabel("Alcohol Content")
plt.ylabel("pH Level")
plt.legend()
plt.show()
```

```python
# Plot Local Outlier Factor outliers
plt.scatter(X[:, 0], X[:, 1], c=np.where(y_pred_lof == -1, 'red', 'pink'), label='Local Outlier Factor')
plt.title("Outlier Detection using Local Outlier Factor on WINE Dataset")
plt.xlabel("Alcohol Content")
plt.ylabel("pH Level")
plt.legend()
plt.show()
```
